# **Introdução**

Antes da criação da Bovespa em 1968, o cenário do mercado de ações no Brasil era fragmentado, com várias bolsas de valores regionais pelo país, cada uma com suas próprias regras. As principais eram as bolsas de São Paulo, Rio de Janeiro e Porto Alegre. A negociação de ações era menos frequente e com menor volume, o que tornava o mercado menos líquido e mais volátil. O acesso era restrito, especialmente para pequenos investidores, e as informações sobre empresas e o mercado eram menos acessíveis, dificultando a tomada de decisões. A regulamentação era menos rigorosa, o que aumentava a insegurança jurídica e o risco de fraudes. Em 1968, o governo lançou o Plano Trienal, buscando modernizar a economia. A criação da Bovespa foi parte desse plano, visando democratizar o acesso ao capital e estimular o investimento privado.

Na década de 70, o Ibovespa experimentou um crescimento acelerado, impulsionado pelo Plano Nacional de Desenvolvimento (PND) e pelo Milagre Econômico. No entanto, a crise do petróleo em 1973 e a instabilidade política resultaram em quedas significativas no final da década. Nos anos 80, o índice enfrentou grande volatilidade devido à crise da dívida externa e à hiperinflação. As recuperações temporárias foram observadas com os Planos Cruzado e Bresser. Na década de 90, com a implementação do Plano Real, o Ibovespa atingiu níveis recordes, impulsionado pela estabilização da economia brasileira. No novo milênio, crises como a bolha da internet em 2001 e a crise financeira global de 2008 impactaram o índice. Em 2008, o Ibovespa perdeu mais de 30% de seu valor. Após a crise de 2008, o Ibovespa se recuperou gradualmente, impulsionado pelo crescimento econômico e pelo aumento do investimento estrangeiro. A pandemia da COVID-19 em 2020 causou instabilidade nos mercados globais, com o Ibovespa sofrendo quedas expressivas, mas recuperando-se posteriormente com medidas de estímulo governamental e expectativas de retomada econômica.

O Erro Quadrático Médio (MSE) é uma forma de medir o quão bem nosso modelo está se ajustando aos dados. Para calculá-lo, primeiro pegamos a diferença entre as previsões feitas pelo modelo e os valores reais, elevamos essas diferenças ao quadrado e então calculamos a média desses quadrados. Quanto menor o MSE, melhor o modelo está ajustando os dados, indicando que as previsões estão mais próximas dos valores reais.

O Erro Médio Absoluto (MAE) é outra medida de precisão das previsões do modelo. Aqui, ao invés de elevar as diferenças ao quadrado, simplesmente pegamos o valor absoluto delas e então calculamos a média. Isso nos dá uma ideia da precisão média das previsões em relação aos valores reais.

Por fim, a Raiz do Erro Quadrático Médio (RMSE) é uma variante do MSE. Calculamos o RMSE tomando a raiz quadrada do MSE. Essa métrica nos dá uma medida do erro médio das previsões em termos da mesma unidade da variável de resposta. O RMSE é bastante utilizado para avaliar a precisão de modelos de previsão, pois nos fornece uma compreensão direta do erro médio das previsões em uma escala semelhante aos valores reais.

Inicialmente foi desenvolvido a função load_and_clean_data() para ajustar os formatos númericos dos dados.

```python
def load_and_clean_data(filepath, sep=';', encoding='ISO-8859-1', decimal=',', thousands='.', header=1, skiprows=2):
    data = pd.read_csv(filepath, sep=sep, encoding=encoding, decimal=decimal, thousands=thousands, header=header, skiprows=skiprows)
    return data
```

Desenvolveu-se a função convert_month() para auxiliar nos ajustes das datas nos dados do IPCA.

```python
def convert_month(month):
    month_dict = {
        'JAN': '01', 'FEV': '02', 'MAR': '03', 'ABR': '04',
        'MAI': '05', 'JUN': '06', 'JUL': '07', 'AGO': '08',
        'SET': '09', 'OUT': '10', 'NOV': '11', 'DEZ': '12'
    }
    if isinstance(month, int):
        return f"{month:02d}"
    return month_dict.get(month.upper(), '01')
```

Desenvolveu-se a função prepare_time_series() para processar e preparar os dados para serem utilizados como série temporal (data como índice).

```python
def prepare_time_series(data, date_col, value_col, year_col=None, month_col=None, date_format='%m/%Y'):
    try:
        if year_col and month_col:
            data['Date'] = data[month_col].apply(convert_month) + '/' + data[year_col].astype(str)
            data = data[['Date', value_col]].copy()
        # Filtrar linhas inválidas
        data = data[pd.to_datetime(data['Date'], format=date_format, errors='coerce').notna()]
        data['Date'] = pd.to_datetime(data['Date'], format=date_format)
        data.set_index('Date', inplace=True)
        data[value_col] = data[value_col].replace({',': '.'}, regex=True).astype(float)
    except KeyError as e:
        print(f"Erro: A coluna {e} não foi encontrada nos dados.")
        print("Colunas disponíveis:", data.columns)
        raise
    except Exception as e:
        print(f"Erro ao preparar a série temporal: {e}")
        raise
    return data
```

Esse aumento representa uma maior inserção de trabalhadores no mercado de trabalho, resultando no aumento da demanda no mercado de bens e serviços, ou seja, forçando o aumento da inflação. Isso representa que as empresas não estão conseguindo ter ganhos de produtividade.

Para alcançar o objetivo do projeto, ou seja, realizar previsões do índice Ibovespa através da análise da série temporal, é importante que a série
seja estacionária. Para que uma série seja estacionária, é necessário que suas estatísticas não mudem de forma significativa ao longo do tempo. Dos gráficos, percebe-se, visualmente, que os dados dos três gráficos não são estacionários, pois os gráficos do IPCA e Ibovespa apresentam tendência de crescimento e a Selic apresenta uma tendência de sazonalidade. Para ter uma certeza maior nessa análise, realizou-se o Teste Dickey-Fuller e a plotagem dos gráficos da função de autocorrelação (ACF) e da função de autocorrelação parcial (PACF).

Para a aplicação do Teste Dickey-Fuller e a plotagem dos gráficos, desenvolveu-se as funções adfuller_test() e plot_acf_pacf(), respectivamente.


```python
# Função para realizar o teste de Dickey-Fuller
def adfuller_test(series, signif=0.05):
    result = adfuller(series.dropna())
    output = {'ADF Statistic': result[0], 'p-value': result[1], 'Critical Values': result[4]}
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])
    if result[1] <= signif:
        print("A série é estacionária")
    else:
        print("A série não á estacionária")
    return output
```

```python
# Função para plotar ACF e PACF
def plot_acf_pacf(series, title=''):
    fig, ax = plt.subplots(2, 1, figsize=(12, 8))
    plot_acf(series.dropna(), lags=20, ax=ax[0])
    ax[0].set_title(f'Autocorrelation Function (ACF) - {title}')
    plot_pacf(series.dropna(), lags=20, ax=ax[1])
    ax[1].set_title(f'Partial Autocorrelation Function (PACF) - {title}')
    plt.tight_layout()
```

Aplicações das funções adfuller_test() e plot_acf_pacf() para a série Selic.

```python
# Realizar a análise da série SELIC
print("Teste de Dickey-Fuller para a série SELIC:")
selic_adf_result = adfuller_test(selic_data['SELIC'])
plot_acf_pacf(selic_data['SELIC'], title='SELIC')
```

Para auxiliar na análise visual dos dados, utilizou-se a função tspplot() do Livro Introdução às Séries Temporais.

```python
def tspplot(ts=None,label=None,title=None,ax=None,linestyle='solid',alpha=1,lw=1,nr_xticks=None,nr_yticks=None,xlabel=None,set_major=None):

  import matplotlib.ticker as ticker

  if ax is None:
    fig, ax = plt.subplots()

  if label is not None:
    ax.plot(ts, label=label, linestyle=linestyle, lw=lw)
  else:
    ax.plot(ts, linestyle=linestyle,lw=lw,alpha=alpha)

  if title is not None:
    ax.set_title(title)

  if xlabel is not None:
    ax.set_xlabel(xlabel)

  if nr_xticks is not None:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nr_xticks))

  if nr_yticks is not None:
    ax.yaxis.set_major_locator(ticker.MaxNLocator(nr_yticks))

  if label is not None:
    plt.legend()

  if set_major is not None:
    ax.xaxis.set_major_locator(set_major)
  plt.tight_layout()

  return
```

Para a análise, separou-se os dados em treinamento e teste. Adotou-se o período dos últimos 12 meses como teste, conforme abaixo:

```python
test_size = 12
train_size = len(data) - test_size

train = data.IBOVESPA[:train_size]
test = data.IBOVESPA[train_size:]
```

Antes de fazer a previsão dos valores do índice Ibovespa, realizou-se a análise dos resíduos do modelo ARIMA, conforme abaixo:

```python
model=ARIMA(train, order=(4, 1, 0))

residuo = model.fit()
residuo_std = (residuo.resid-residuo.resid.mean())/np.sqrt(residuo.resid.var())
ts=data

fig,axs = plt.subplots(2, 2,figsize=(14,5))

axs[0,0].scatter(ts.index[1:train_size], residuo_std[1:train_size])
axs[0,0].axhline(y=0, color='b', linestyle='--')
axs[0,0].set_title('Standardized residuals')

x=plot_acf(residuo.resid, lags=20, ax=axs[0,1], title="ACF- resíduos");

axs[1,0].hist(residuo_std[1:train_size], density=True);
x=np.sort(np.random.normal(0, 1, 1000))
axs[1,0].plot(x, stats.norm.pdf(x, 0, 1))
x1=sm.qqplot(residuo_std[1:train_size], line ='45', ax=axs[1,1])
```

Dos resultados acima, verifica-se que a informação presente na série Ibovespa foi considerado no modelo ARIMA, pois, no gráfico ACF os valores são todos próximos de zero e os resíduos apresentam uma distribuição quase normal.

Após fazer essa verificação, realizou-se a previsão dos últimos dozes meses da série temporal. Essa ação é necessária para verificar a acurácia do modelo.

```python
forecast = model.fit().get_forecast(12)
forecast.predicted_mean

# Previsões com intervalos de confiança de 95%
fc_series = pd.Series(forecast.predicted_mean, index=test.index)
lower_series = pd.Series(forecast.conf_int(alpha = 0.05)["lower IBOVESPA"], index=test.index)
upper_series = pd.Series(forecast.conf_int(alpha = 0.05)["upper IBOVESPA"], index=test.index)

fc_series.head()
```

```python
plt.figure(figsize=(15,5))

plt.plot(train, label='treino',linestyle='solid',alpha=1,lw=1)
plt.plot(test, label='teste',linestyle='solid',alpha=1,lw=1)
plt.plot(fc_series, label='previsão',linestyle='dashed',alpha=1,lw=2)
plt.fill_between(lower_series.index, lower_series, upper_series,
                 color='k', alpha=.15)
plt.title('Previsão vs Atual')
plt.legend(loc='upper left', fontsize=8)

plt.show()
```

Dos resultados acima, verifica-se os primeiros cinco valores previstos, dentro do modelagem, e o gráfico apresentando os resultados obtidos. As previsões foram realizadas com um intervalo de precisão de 95%.

Dos resultados obtidos, obtem-se os valores das métricas de acurária do modelo:

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    return({'mape':mape, 'me':me, 'mae': mae,
            'mpe': mpe, 'rmse':rmse})

forecast_accuracy(forecast.predicted_mean, test.values)
```
