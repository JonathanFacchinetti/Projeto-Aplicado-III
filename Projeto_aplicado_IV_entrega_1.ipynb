# **Resumo**

Este projeto tem o objetivo de estudar a influência da taxa de inflação e a SELIC no mercado acionário brasileiro, representado pela IBOVESPA. Com base nesse estudo será possível realizar previsões de como o consumo e as políticas monetárias afetam o mercado bens e serviços brasileiro. O estudo será realizado através de Séries Temporais da taxa de inflação, taxa SELIC e o índice IBOVESPA. Será realizado a decomposição das Séries Temporais em Tendência, Ciclos, Sazonalidade e Resíduos para serem utilizados nas previsões. Para a aplicação das técnicas mencionadas, será utilizado a linguagem de programação Python e as suas bibliotecas, principalmente o stats models.  


# **Fonte dos Dados**

Os dados utilizados são a evolução mensal do índice IBOVESPA (https://www.b3.com.br/pt_br/market-data-e-indices/indices/indices-amplos/indice-ibovespa-ibovespa-estatisticas-historicas.htm), a taxa SELIC (https://www3.bcb.gov.br/sgspub/consultarvalores/consultarValoresSeries.do?method=getPagina) e a taxa de inflação IPCA (https://www.ibge.gov.br/estatisticas/economicas/precos-e-custos/9256-indice-nacional-de-precos-ao-consumidor-amplo.html?=&t=series-historicas) do IBGE. Os dados são formados pelo mês e seus respectivos valores. O IBOVESPA é o indicador mais importante e seu valor é resultado do desempenho das ações negociadas na B3. A SELIC é a taxa básica de juros e serve como base para todas as taxas de juros no país. O Índice Nacional de Preços ao Consumidor Amplo (IPCA) é um dos índices de inflação mais importantes utilizado no país e seu valor é calculado pelo IBGE.
